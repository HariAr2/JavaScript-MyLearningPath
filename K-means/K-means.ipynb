## K-Means Clustering

```python
x1 = df.loc[::,["Age","Spending Score (1-100)"]].values
from sklearn.cluster import KMeans
wcss = [] #within cluster sum of squares

for k in range (1,11):
    kmeans = KMeans(n_clusters=k,random_state=7)
    kmeans.fit(x1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,linewidth = 2,color = "red",marker = 11)
```

## Cluster Centroids or Centres

```python
print(kmeans.cluster_centers_)
```

```python
kmeans = KMeans(n_clusters=5)
cluster = kmeans.fit_predict(x1)
print(kmeans.cluster_centers_)
```

## Visualization of Result

```python
plt.scatter(x1[:,0],x1[:,1],c=kmeans.labels_,cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker="*",color="black",s=300)
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("K-Means clustering for Age VS Spending Score")
```
